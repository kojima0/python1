```python
y1 = np.random.rand(11)
y2 = np.random.rand(11)

fig = plt.figure()

sp1 = fig.add_subplot(1, 2, 1)
sp1.plot(y1)

sp2 = fig.add_subplot(1, 2, 2)
sp2.plot(y2)
```

```python
# サンプルデータ
x = np.array([20, 30, 40, 50, 60])
y = np.array([30, 38, 41, 49, 62])

# 近似直線の式の　a と b が入ったタプルを得る
p = np.polyfit(x, y, 1)

# 一次関数の式のオブジェクトを生成する
f = np.poly1d(p)

# 散布図と近似直線を描く
plt.scatter(x, y)
plt.plot(x, f(x))
```
